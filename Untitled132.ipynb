```python
import pandas as pd
import numpy as np
import wbgapi as wb
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
country_codes = ["LTU","LUX", 'LVA', 'MAC', 'MAF', 'BDI', 'BFA', 'CAF', 'BEN', 'BLZ'] 

country_names = {"LTU": 'Lithuania', 
                 "LUX": 'Luxembourg', 
                 "LVA": 'Latvia', 
                 "MAC": 'Macau', 
                 "MAF": 'Saint Martin', 
                 "BDI": 'Burundi',
                 "BFA": 'Burkina Faso', 
                 "CAF": 'Central African Republic',
                 "BEN": 'Benin', 
                 "BLZ": 'Belize'}

indicator_ids = ["AG.LND.ARBL.ZS", "AG.LND.FRST.ZS"]
```

```python
df  = wb.data.DataFrame(indicator_ids, country_codes, mrv=20) # most recent 10 years
df.reset_index(inplace = True)
df
```

```python
df.replace(['AG.LND.ARBL.ZS', 'AG.LND.FRST.ZS'], ['Forest land', "Arable land"], inplace = True)
df.rename(columns = {'economy':'Country', 'series':'Land %'}, inplace = True)
```

```python
df['Country'] = df['Country'].map(country_names)
df.head()
```

```python
df =pd.pivot_table(df,index=["Country"],values=['YR2001', 'YR2002', 'YR2003', 'YR2004',
       'YR2005', 'YR2006', 'YR2007', 'YR2008', 'YR2009', 'YR2010', 'YR2011',
       'YR2012', 'YR2013', 'YR2014', 'YR2015', 'YR2016', 'YR2017', 'YR2018',
       'YR2019', 'YR2020'], columns = ['Land %'])
df = df.T.reset_index()
```

```python
df = df.rename(columns = {'level_0':'Year'})
```

```python
df['Year'] = df['Year'].str.replace('YR', '')
df["Year"] = df['Year'].astype(int)
```

```python
df.head(3)
```

```python

plt.figure(figsize=(20, 20))

for e, column in enumerate(['Lithuania', 'Luxembourg', 'Latvia']):
    plt.subplot(4, 3, e + 1)
    sns.barplot(x='Land %',y=column,data=df)
    plt.ylabel('Frequency')
```

```python

plt.figure(figsize=(20, 20))

for e, column in enumerate(['Belize', 'Benin', 'Burundi']):
    plt.subplot(4, 3, e + 1)
    sns.barplot(x='Land %',y=column,data=df)
    plt.ylabel('Frequency')
```
