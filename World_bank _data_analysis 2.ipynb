#### Import Required Python Packages

```python
import pandas as pd
import numpy as np
import wbgapi as wb
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

#### Read data using wbgapi API

```python
country_codes = ["LTU","LUX", 'LVA', 'BRA', 'MAF', 'BDI', 'BFA', 'CAF', 'BEN', 'BLZ'] 


### We have selected 5 high income and 5 low income countries from world bank website
country_names = {"LTU": 'Lithuania', 
                 "LUX": 'Luxembourg', 
                 "LVA": 'Latvia', 
                 "BRA": 'Brazil', 
                 "MAF": 'Saint Martin', 
                 "BDI": 'Burundi',
                 "BFA": 'Burkina Faso', 
                 "CAF": 'Central African Republic',
                 "BEN": 'Benin', 
                 "BLZ": 'Belize'}

indicator_ids = ["AG.LND.ARBL.ZS", "AG.LND.FRST.ZS"]
```

```python
df  = wb.data.DataFrame(indicator_ids, country_codes, mrv=30) # most recent 10 years
df.reset_index(inplace = True)
df
```

```python
def data_preprocessing(data):
    
    """
    This code will do all the required pro-processing in  data to get it into the format
    that is required to perform analysis. We need Year and country names as columns.
    This function will make use of pivot table as well.
    """
    data.replace(['AG.LND.ARBL.ZS', 'AG.LND.FRST.ZS'], ['Forest land', "Arable land"], inplace = True)
    data.rename(columns = {'economy':'Country', 'series':'Land'}, inplace = True)
    data['Country'] = data['Country'].map(country_names)
    data =pd.pivot_table(data,index=["Country"],values=['YR1991', 'YR1992', 'YR1993', 'YR1994', 'YR1995',
       'YR1996', 'YR1997', 'YR1998', 'YR1999', 'YR2000', 'YR2001', 'YR2002',
       'YR2003', 'YR2004', 'YR2005', 'YR2006', 'YR2007', 'YR2008', 'YR2009',
       'YR2010', 'YR2011', 'YR2012', 'YR2013', 'YR2014', 'YR2015', 'YR2016',
       'YR2017', 'YR2018', 'YR2019', 'YR2020'], columns = ['Land'])
    data = data.T.reset_index()
    data = data.rename(columns = {'level_0':'Year'})
    data['Year'] = data['Year'].str.replace('YR', '')
    data["Year"] = data['Year'].astype(int)
    return data

df = data_preprocessing(df)
df.head()
```

```python
def describe(column, label):
    """
        Pandas describe function to get statistical insights of numerical data.
    """
    return df.loc[df[column] == label].describe().T.style.background_gradient(cmap='Set2',low =0.4,high=0.1,axis=0)

# For Arable Land

describe("Land", "Arable land")
```

```python
## USe of same above function for Forest lands summary

describe("Land", "Forest land")
```

__Inference:__  The table above depicts various statistical measures of forest and arable land area in high- and low-income countries. Forests cover a large portion of the land in high-income countries like Latvia, Lithuania, and Luxembourg.

```python
## Because bar plot are suitable to compare and visualize continusous numerical and cateogrical features
def barplot(x, y):
    
    """"
    Developing bar plots to compare the Forest and Arable Land percentage in 
    High income and low income countries
    """
    plt.figure(figsize=(20, 20))

    for e, column in enumerate(y):
        plt.subplot(4, 3, e + 1)
        sns.barplot(x=x ,y=column, data=df)
        plt.ylabel(column,fontsize=14)
        
## Develop graph for 3 high income countries    
barplot('Land', ['Lithuania', 'Luxembourg', 'Latvia'])
```

```python
## use same above function for Low income countries

barplot('Land', ['Belize', 'Benin', 'Burundi'])
```

__Inference:__ We can observe from the above plot that Lithuania has more forest land (more than 30%) as compared to arable land. Low-income countries' land records, on the other hand, clearly indicate how their governments have mismanaged their lands. Belize uses 60% of its land for agriculture and has less than 5% of its area for forest. 

```python
## Because scatter plot are suitable to compare and visualize two numerical features, we'll use it's hue functio as well to 
## differentiate between two different categories and get data for each in different color.

def scatterplot(x, y):
    
    """"
    Developing scatterplots to compare the development or decline in forest land over time
    in High and low income countries.
    """

    plt.figure(figsize=(20, 20))

    for e, column in enumerate(y):
        plt.subplot(4, 3, e + 1)
        sns.scatterplot(data=df, x=x, y=column, hue="Land")
        plt.xticks(rotation = 45,fontsize=14)
        plt.ylabel(column,fontsize=14)

## Develop scatterplot for 3 high income countries    
scatterplot("Year", ['Lithuania', 'Luxembourg', 'Latvia'])
```

__Inference:__ We can observe from the above figure that frest lands are increasing over time in developed countries while Arable lands have decline at some points.

```python
## Lineplots are suitable as well to plot the correlation between two numerical feautres.

def lineplot(x, y):
    
    """"
    Developing scatterplots to compare the development or decline in forest land over time
    in High and low income countries.
    """

    plt.figure(figsize=(20, 20))

    for e, column in enumerate(y):
        plt.subplot(4, 3, e + 1)
        sns.lineplot(data=df, x=x, y=column, hue="Land", lw=2)
        plt.xticks(rotation = 45,fontsize=14)
        plt.ylabel(column,fontsize=14)

## Develop scatterplot for 3 high income countries    
lineplot("Year", ['Lithuania', 'Luxembourg', 'Latvia'])
```

__Inference:__ The graph above indicates that forest land in three of our high-income countries has been steadily increasing over time. While we can observe from that Low income countries have continual decline in their Forest land over time.
